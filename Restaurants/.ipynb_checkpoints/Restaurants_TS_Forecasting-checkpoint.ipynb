```python
train_df = pd.merge(train_df,test_df[['City','Restaurant']].drop_duplicates(), on=['City','Restaurant'],how='inner')
train_df['Date'] = pd.to_datetime(train_df['Date'], dayfirst=True, format='%d-%m-%y')
```

```python
train_df
```

```python
city_df = pd.read_csv(data_path)
city_df['Date'] = pd.to_datetime(city_df['Date'], dayfirst=True, format='%d-%m-%y')
```

```python
city_df.info()
```

#### Fill missing 'Unemployment' values by city mean

```python
city_df['Unemployment'] = city_df.groupby('City')['Unemployment'].transform(lambda x: x.fillna(x.mean()))
```

```python
city_df.info()
```

### Analyse data for each unique restaurant

```python
size = train_df.groupby(['Restaurant','City']).size()

print(f'Min value: {min(size)}, max : {max(size)} ')
print(f'{len(size[size < 100])} restaurants have less then 100 values')
```

Not all restaurants have full data, need to fill every missing data, so each restaurant have same amount of data

### Create time series for each unique restaurant

Each df has to be indexed with dates, have exactly 100 values (NaN's are filled with mean)

```python
def get_ts_df(city_id, restaurant_id,with_exog=False, df=train_df, city_df=city_df):
    ts_df = df[(df['City'] == city_id) & (df['Restaurant'] == restaurant_id)].set_index('Date')
       
    idx_range = pd.date_range(start = min(df['Date']), end = max(df['Date']),freq='7D')
    ts_df = ts_df.reindex(idx_range)
    
    ts_df['Weekly_Sales'] = ts_df['Weekly_Sales'].fillna( ts_df['Weekly_Sales'].mean() )
    
    if with_exog:
        city_info = city_df[ city_df['City'] == city_id ].set_index('Date')
        ts_df = pd.merge(ts_df, city_info,left_index=True, right_index=True, on=['City','IsHoliday'])
        ts_df.update(city_info['IsHoliday'])         
    else:
        ts_df = ts_df.drop(['IsHoliday'],axis=1)
    
    ts_df = ts_df.drop(['City','Restaurant'],axis=1)
    return ts_df
```

```python
ts_df = get_ts_df(1,2)
```

## Models

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
 ts_train, ts_test = train_test_split(ts_df, test_size=0.2, shuffle=False)
```

```python
def plot_dif(predicted_x,predicted_y, actual):
    fig = plt.figure(facecolor='w', figsize=(16,9))
    ax = fig.add_subplot(111)
    ax.plot(actual.index,actual['Weekly_Sales'],label='observed')
    
    ax.plot(predicted_x,predicted_y,label='predicted')
    
    ax.legend()
    ax.set_xlabel('Date')
    ax.set_ylabel('Restaurant Sales')
    plt.title('Observed vs Predicted')
    plt.show()
```

### FB Prophet

```python
from fbprophet import Prophet
```

```python
def to_fb_format(df):
    return pd.DataFrame({'ds':df.index,'y':df['Weekly_Sales']}).reset_index(drop=True)
```

```python
ts_train_fb = to_fb_format(ts_train)
ts_test_fb = to_fb_format(ts_test)
```

```python
m = Prophet(yearly_seasonality=True)
m.fit(ts_train_fb)
```

```python
fb_pred = m.predict(ts_test_fb)
```

```python
fig1 = m.plot(fb_pred)
```

```python
fig2 = m.plot_components(fb_pred)
```

```python
plot_dif(fb_pred['ds'],fb_pred['yhat'], ts_df)
```

```python
fb_rmse = np.sqrt(mean_squared_error(ts_test_fb['y'],fb_pred['yhat']))
print(f'RMSE: {fb_rmse}')
```

## Stats models

```python
ts_df = get_ts_df(1,2,with_exog=True)

endog = ts_df['Weekly_Sales']

exog = ts_df.drop(['Weekly_Sales'], axis=1)
exog['IsHoliday'] = exog['IsHoliday'].astype(int)

endog_train, endog_test, exog_train, exog_test = train_test_split(endog, exog, test_size=0.2, shuffle=False)
```

### Exp smoothing

```python
model = sm.tsa.ExponentialSmoothing(endog_train, seasonal_periods=12, trend='add', seasonal='mul',damped=True)
```

```python
results = model.fit()
```

```python
exp_pred = results.predict(start=min(endog_test.index),end=max(endog_test.index)) 
```

```python
plot_dif(exp_pred.index,exp_pred.values, ts_df)
```

```python
exp_rmse = np.sqrt(mean_squared_error(endog_test,exp_pred.values))
print(f'RMSE: {exp_rmse}')
```

### ARIMA

```python
import itertools
```

```python
def grid_search_arima(train,pdq_range=3,exog=None):
    p = d = q = range(pdq_range)
    pdq = list(itertools.product(p, d, q))   
    best_rmse = np.inf
    best_params = ()

    for param in pdq:
            try:
                model = sm.tsa.ARIMA(train,order=param, exog=exog)
                results = model.fit()
                pred = results.predict(start=min(endog_test.index),end=max(endog_test.index),exog=exog_test)
                rmse = np.sqrt(mean_squared_error(endog_test,pred))
                if rmse < best_rmse:                    
                    best_params = param
                    best_rmse= rmse                    
            except:
                continue
    print('The best ones are:', best_rmse, best_params)
```

```python
grid_search_arima(endog_train,exog=exog_train)
```

```python
model = sm.tsa.ARIMA(endog_train,order=(1,0,2),exog=exog_train)
```

```python
results = model.fit()
```

```python
arima_pred = results.predict(start=min(endog_test.index),end=max(endog_test.index),exog=exog_test) 
```

```python
plot_dif(arima_pred.index,arima_pred.values, ts_df)
```

```python
arima_rmse = np.sqrt(mean_squared_error(endog_test,arima_pred.values))
print(f'RMSE: {arima_rmse}')
```

### SARIMAX

```python
def grid_search_sarimax(train,pdq_range=3,periods=12,exog=None):
    p = d = q = range(pdq_range)
    pdq = list(itertools.product(p, d, q))
    seasonal_pdq = [(x[0], x[1], x[2], periods) for x in pdq]
    best_rmse = np.inf
    best_params = ()

    for param in pdq:
        for param_seasonal in seasonal_pdq:
            try:
                model = sm.tsa.statespace.SARIMAX(train,
                                                order=param,
                                                exog=exog,
                                                seasonal_order=param_seasonal,
                                                time_varying_regression=True,
                                                mle_regression=False,
                                                trend='t')
                results = model.fit()
                pred = results.predict(start=min(endog_test.index),end=max(endog_test.index),exog=exog_test)
                rmse = np.sqrt(mean_squared_error(endog_test,pred))
                if rmse < best_rmse:                    
                    best_params = (param,param_seasonal)
                    best_rmse= rmse
            except:
                continue
    print('The best ones are:', best_rmse, best_params)
```

```python
grid_search_sarimax(endog_train,exog=exog_train)
```

```python
model = sm.tsa.statespace.SARIMAX(endog_train,
                                            order=(1,1,2),
                                            seasonal_order=(2,0,2,12),
                                              exog=exog_train,
                                  time_varying_regression=True,
                                  mle_regression=False,
                                  trend='t')
```

```python
results = model.fit()
```

```python
fig = results.plot_diagnostics()
```

```python
sarimax_pred = results.predict(start=min(endog_test.index),end=max(endog_test.index),exog=exog_test) 
```

```python
plot_dif(sarimax_pred.index,sarimax_pred.values, ts_df)
```

```python
sarimax_rmse = np.sqrt(mean_squared_error(endog_test,sarimax_pred.values))
print(f'RMSE: {sarimax_rmse}')
```

```python
def plot_models(actual,models_pred):
    fig = plt.figure(facecolor='w', figsize=(16,9))
    ax = fig.add_subplot(111)
    ax.plot(actual.index,actual.values,label='observed')
    
    for x,y,name in models_pred:
        ax.plot(x,y,label=name)
    
    ax.legend()
    ax.set_xlabel('Date')
    ax.set_ylabel('Restaurant Sales')
    plt.title('Compare models predictions')
    plt.show()
```

```python
models_pred = [ (fb_pred['ds'],fb_pred['yhat'],'Prophet'),
 (exp_pred.index,exp_pred.values,'Exp. smooting'),
 (arima_pred.index,arima_pred.values,'ARIMA'),
 (sarimax_pred.index,sarimax_pred.values,'SARIMAX')
] 
```

```python
plot_models(endog_test,models_pred)
```

## Predict

Prophet is the best model when comparing RMSE, so it will be used as final model for test predictions

```python
%%capture

predictions = {}
for _,(city,restaurant) in train_df[['City','Restaurant']].drop_duplicates().iterrows():
    ts_df = get_ts_df(city,restaurant)
    ts_df = to_fb_format(ts_df)
    m = Prophet(yearly_seasonality=True)
    m.fit(ts_df)
    future = m.make_future_dataframe(periods=50,freq='7d')
    forecast = m.predict(future)
    predictions[(city,restaurant)] = forecast
```

```python
pred_column = []
for _, row in test_df.iterrows():
    try:
        pred = predictions[row['City'],row['Restaurant']]
    except KeyError:
        # There are some restaurants in test that are not included in original train         
        pred_column.append(None) 
        continue
    pred_column.append(pred[pred['ds'] == row['Date']]['yhat'].values[0]) 
```

```python
test_df['Weekly_Sales'] = pred_column
```

```python
test_df['Weekly_Sales'] = test_df.groupby('City')['Weekly_Sales'].transform(lambda x: x.fillna(x.mean()))
```
